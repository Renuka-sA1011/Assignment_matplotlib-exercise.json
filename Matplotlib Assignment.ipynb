```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
```

# Exercise 2

```python
fig=plt.figure()                          
axes=fig.add_axes([0,0,1,1])               

axes.plot(x,y)                       
axes.set_xlabel('X Axis Values')         
axes.set_ylabel('Y Axis values')
axes.set_title('Graph X Vs Y')
```

```python
fig=plt.figure()                         
ax1 = fig.add_axes([0,0,1,1])      
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])       

ax1.plot(x,y,'r')                       
ax1.set_xlabel('X Axis Values')         
ax1.set_ylabel('Y Axis values')
ax1.set_title('Graph X Vs Y')

ax2.plot(x,y,'b')                       
ax2.set_xlabel('X Axis Values')         
ax2.set_ylabel('Y Axis values')
ax2.set_title('Graph X Vs Y')
```

# Exercise 3

```python
fig=plt.figure()                         
axes3 = fig.add_axes([0,0,1,1])      
axes4 = fig.add_axes([0.2,0.5,0.4,0.4])       

axes3.plot(x,y,'c')                       
axes3.set_xlabel('X Axis Values')         
axes3.set_ylabel('Y Axis values')
axes3.set_title('Graph X Vs Y')

axes4.plot(x,y,'m')                       
axes4.set_xlabel('X Axis Values')         
axes4.set_ylabel('Y Axis values')
axes4.set_title('Graph X Vs Y')
```

```python
plt.subplots(nrows=1, ncols=2)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.plot(x,color='g', linewidth=1)
ax.plot(x,y,'-.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('X and Y')
```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,color='m', linewidth=2.00)
ax.plot(x,z,':')
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_title('X and Z')
```

```python
fig,ax=plt.subplots(figsize=(17,4))
ax.plot(y,color='c', linewidth=2.00)
ax.plot(x,y,'-')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('X and Y')
```
